1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Importando o dataset

df = pd.read_csv('previsao_de_renda.csv')

## Mantendo as variáveis qualitativas e a variável target
qual_df = df[['renda', 'posse_de_veiculo', 'posse_de_imovel']]
qual_df.head()
````

````python
fig, axes = plt.subplots(1, 2)
fig.set_size_inches(16, 9)

sns.pointplot(x='posse_de_veiculo', y='renda', data=qual_df, ax=axes[0], dodge=True, errorbar=('ci', 95))
sns.pointplot(x='posse_de_imovel', y='renda', data=qual_df, ax=axes[1], dodge=True, errorbar=('ci', 95))
````

Ambas as variáveis parecem ser estatisticamente significativas, em ambos os casos, o intervalo de confiança dos dois estão bem distantes, o que indica que a média de renda é diferente para cada um dos casos.
